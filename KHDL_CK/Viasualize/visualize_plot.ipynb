```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\KHDL_CK\Data\patekphilippe_clean_data_pure_train.csv')
print(df.head())
```

```python
# Tính toán ma trận tương quan
correlation_matrix = df.corr()

# Lấy tương quan của các đặc trưng với biến mục tiêu 'price($)'
correlation_with_price = correlation_matrix['price($)'].sort_values(ascending=False)

# In các đặc trưng có giá trị tương quan cao với 'price($)'
print("Các đặc trưng có giá trị tương quan cao với price($):")
print(correlation_with_price)

# Trực quan hóa bằng biểu đồ cột
plt.figure(figsize=(10, 6))
correlation_with_price.drop('price($)').plot(kind='bar')
plt.title('Tương quan của các đặc trưng với price($)')
plt.xlabel('Đặc trưng')
plt.ylabel('Hệ số tương quan')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Biểu diễn số lượng biến 'price($)'
plt.figure(figsize=(10, 6))
df['price($)'].plot(kind='hist', bins=20)
plt.title('Phân phối của price($)')
plt.xlabel('price($)')
plt.ylabel('Số lượng')
plt.tight_layout()
plt.show()
```

```python
# Giá trị lớn nhất của 'price($)'
max_price = df['price($)'].max()
print("Giá trị của 'price($)':", max_price)

# Lấy dòng chứa giá trị lớn nhất của 'price($)'
max_price_row = df[df['price($)'] == max_price]
print("Dòng chứa giá trị lớn nhất của 'price($)':")
print(max_price_row)
```

```python
# Biểu diễn giá trị ngoại lại của biến 'price($)'
plt.figure(figsize=(10, 6))
df.boxplot(column='price($)')
plt.title('Biểu diễn giá trị ngoại lại của price($)')
plt.ylabel('price($)')
plt.tight_layout()
plt.show()
```

```python
# Bỏ các giá trị ngoại lại của biến 'price($)'
df = df[df['price($)'] < 170000]
plt.figure(figsize=(10, 6))
```

```python
# Biểu diễn giá trị ngoại lại của biến 'price($)'
plt.figure(figsize=(10, 6))
df.boxplot(column='price($)')
plt.title('Biểu diễn giá trị ngoại lại của price($)')
plt.ylabel('price($)')
plt.tight_layout()
plt.show()
```

```python
# Biểu diễn bằng scatter plot
plt.scatter(df['Case area(mm2)'], df['price($)'])
plt.title('Biểu diễn giữa year và price($)')
plt.xlabel('year')
plt.ylabel('price($)')
plt.tight_layout()
plt.show()
```

```python
# Đếm số lượng mẫu có 'price($)'] < 180000
count = df[df['price($)'] < 170000].shape[0]
print("Số lượng mẫu có 'price($)'] < 180000:", count)
```
